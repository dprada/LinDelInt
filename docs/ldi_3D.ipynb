```python
import lindelint as ldi
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
n_vertices = 4

lbox=10
max_module = 2.0

vertices = np.random.uniform(-lbox+max_module, lbox-max_module, (n_vertices, 3))
v_vertices = np.random.uniform(-1, 1, (n_vertices, 3))
v_vertices = max_module*v_vertices/np.linalg.norm(np.ravel(v_vertices))
```

```python
vertices[0]=[-5, 5, 0]
vertices[1]=[-5, -5, 0]
vertices[2]=[7, 0, 0]
vertices[3]=[0, 0, 7]
```

```python
n_points = 5000
points = np.random.uniform(-10, 10, (n_points, 3))
```

```python
interpolator = ldi.Interpolator(vertices, v_vertices)
```

```python
va = interpolator.do_your_thing(points)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vertices[:,0], vertices[:,1], vertices[:,2])
ax.plot([vertices[1,0], vertices[2,0]], [vertices[1,1],vertices[2,1]], [vertices[1,2],vertices[2,2]])

ax.scatter(points[va,0], points[va,1], points[va,2], color='r')

ax.axes.set_xlim3d(-10, 10) 
ax.axes.set_ylim3d(-10, 10) 
ax.axes.set_zlim3d(-10, 10) 

plt.show()
```
