# 2 points in 2D

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lbox=10

n_centers = 2
max_module = 2.0

centers = np.random.uniform(-lbox+max_module, lbox-max_module, (n_centers, 2))
v_centers = np.random.uniform(-1, 1, (n_centers, 2))
v_centers = max_module*v_centers/np.linalg.norm(np.ravel(v_centers))
```

```python
n_points = 200
points = np.random.uniform(-10, 10, (n_points, 2))
v_points = np.zeros((n_points, 2))
```

```python
#centers[:,:] = 0.0
#centers[0,0] = -5.0
#centers[1,0] = 5.0

#v_centers[:,:] = 0.0
#v_centers[0,1] = 1.0
#v_centers[1,0] = 1.0
```

```python
plt.rcParams["figure.figsize"] = (6,6)

ax=plt.gca()

for ii in range(n_centers):
    plt.arrow(centers[ii,0], centers[ii,1], v_centers[ii,0], v_centers[ii,1],
              head_width=0.2, head_length=0.4, fc='r', ec='r')

plt.scatter(points[:,0], points[:,1], s=5, c='k')
    
plt.axis('scaled')
plt.xlim(-lbox, lbox)
plt.ylim(-lbox, lbox)

plt.show()
```

```python
v01 = centers[1]-centers[0]
d01 = np.linalg.norm(v01)
u01 = v01/d01
```

```python
for ii in range(n_points):
    v0p = points[ii]-centers[0]
    f = np.dot(v0p, u01)/d01
    if f<0:
        f = 0
    elif f>1:
        f = 1
    v_points[ii,:] = v_centers[0] + (v_centers[1]-v_centers[0]) * f
```

```python
plt.rcParams["figure.figsize"] = (6,6)

ax=plt.gca()

for ii in range(n_centers):
    plt.arrow(centers[ii,0], centers[ii,1], v_centers[ii,0], v_centers[ii,1],
              head_width=0.2, head_length=0.4, fc='r', ec='r')

for ii in range(n_points):
    plt.arrow(points[ii,0], points[ii,1], v_points[ii,0], v_points[ii,1],
          head_width=0.2, head_length=0.4, fc='k', ec='k')
    
plt.axis('scaled')
plt.xlim(-lbox, lbox)
plt.ylim(-lbox, lbox)

plt.show()
```
